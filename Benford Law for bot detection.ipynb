```python
import numpy as np
import pandas as pd
from benford_law import Benford
```

```python
bot_data = pd.read_csv('bots_data.csv')

bot_data2 = pd.read_csv('FYP/datasets_full/fake_followers/users.csv')
botss_data= pd.read_json("/Users/Envy1/Desktop/botwiki-2019/botwiki-2019_tweets.json")
bd=pd.json_normalize(botss_data.user)
human_data = pd.read_json("/Users/Envy1/Desktop/verified-2019/verified-2019_tweets.json")
hd=pd.json_normalize(human_data.user)
```

```python
nonbot_data = pd.read_csv('Humans_data.csv')
nonbot_data
nonbot_data2 = pd.read_csv('FYP/datasets_full/genuine_accounts/users.csv')
```

```python
bot_attr = bot_data[
  ['followers_count', 'friends_count', 'listedcount', 'favourites_count', 'statuses_count','id']]
bot_attr2 = bot_data2[
  ['followers_count', 'friends_count', 'listed_count', 'favourites_count', 'statuses_count','id']]
bot_label = bot_data[['bot']]
bot_attr
```

```python
nonbot_attr = nonbot_data[
  ['followers_count', 'friends_count', 'listedcount', 'favourites_count', 'statuses_count','id']]
nonbot_attr2 = nonbot_data2[
  ['followers_count', 'friends_count', 'listed_count', 'favourites_count', 'statuses_count','id']]
nonbot_attr
```

```python

print("\n Benford law for bot followers count \n")
b2=Benford(np.concatenate((bd.followers_count,bot_attr.followers_count,bot_attr2.followers_count)))
b2.benford_analysis()
b2.plot()
print("\n Benford law for nonbot followers count \n")
b1= Benford(np.concatenate((bd.followers_count,nonbot_attr.followers_count,nonbot_attr.followers_count,nonbot_attr2.followers_count)))
b1.benford_analysis()
b1.plot()
```

```python
print("\n Benford law for bot friends count \n")
b2=Benford(np.concatenate((bd.friends_count,bot_attr.friends_count,bot_attr2.friends_count)))
b2.benford_analysis()
b2.plot()
print("\n Benford law for nonbot friends count \n")
b1= Benford(np.concatenate((hd.followers_count,nonbot_attr.friends_count,nonbot_attr2.friends_count)))
b1.benford_analysis()
b1.plot()
np.concatenate((bd.friends_count,bot_attr.friends_count,bot_attr2.friends_count)).size
```

```python
print("\n Benford law for bot listed count \n")
b2=Benford(np.concatenate((bd.listed_count,bot_attr.listedcount,bot_attr2.listed_count)))
b2.benford_analysis()
b2.plot()
print("\n Benford law for nonbot listed count \n")
b1= Benford(np.concatenate((hd.listed_count,nonbot_attr.listedcount,nonbot_attr2.listed_count)))
b1.benford_analysis()
b1.plot()
np.concatenate((hd.listed_count,nonbot_attr.listedcount,nonbot_attr2.listed_count)).size
```
